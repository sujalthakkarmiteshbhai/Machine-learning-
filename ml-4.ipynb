```python
from numpy import sign
def intercept_slop(x,y):
    n=len(x)
    sig_x=np.sum(x)
    sig_y=np.sum(y)
    sign_xy=np.sum(x*y)
    sig_x2=np.sum(x**2)
    m = (n*sign_xy-sig_x*sig_y)/(n*sig_x2-sig_x**2)
    c = (sig_y-m*sig_x)/n
    return m,c

m,c = intercept_slop(df["Height(Inches)"],df["Weight(Pounds)"])
print(m,c)
```
